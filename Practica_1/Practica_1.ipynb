## Cinemática de los robots móviles sobre ruedas
Existen varios tipos de modelos cinemáticos:

###  La cinemática interna
Explica la relación entre las variables internas del sistema por ejemplo, rotación de la rueda y el movimiento del robot.

### La cinemática externa
Describe la posición y orientación del robot de acuerdo con
algún marco de coordenadas de referencia.

### Cinemática directa y cinemática inversa
<div style="text-align: justify">La cinemática directa describe
estados del robot en función de sus entradas (velocidades de las ruedas, movimiento de las articulaciones,
volante, etc.). A partir de la cinemática inversa se puede diseñar un movimiento
planificación, lo que significa que las entradas del robot se pueden calcular para un
secuencia de estado del robot deseada. </div> 

##  Robot de accionamiento diferencial

<div style="text-align: justify">La Robot Movil diferencial es un mecanismo de conducción muy simple, con aplicaciones prácticas, especialmente para robots móviles pequeños. Estos Robots con esta configuracion
suelen tener una o más ruedas giratorias (Locas) para sostener el vehículo y evitar
inclinación. Posee dos  ruedas principales las cuales se colocan en un eje común. La velocidad de
cada rueda está controlada por un motor independiente. </div> 

<!-- ![Alt Text](image path "title") -->
![Alt Text](4.png "Robot Diferencial")

<div style="text-align: justify">De acuerdo con la figura, las variables de entrada (control) son la velocidad de la rueda derecha $v_r(t)$ y la velocidad de la rueda izquierda $v_l(t)$. El significado de las demas variables en la figura son los siguientes: $r$ es el radio de la rueda, $L$ es la distancia entre las ruedas, $R(t)$ es el radio instantáneo de la trayectoria de conducción del vehículo (distancia entre el centro del vehículo y Punto ICR (centros instantáneos de rotación)). En cada instancia de tiempo, ambas ruedas tienen la misma velocidad angular $\omega(t)$ alrededor del ICR como se muestra a continuacion </div>

$$\omega(t)=\frac{v_l(t)}{R(t)-\frac{L}{2}}$$

$$\omega(t)=\frac{v_r(t)}{R(t)+\frac{L}{2}}$$

De donde $\omega(t)$ y $R(t)$ se expresan a continuación:

$$\omega(t)=\frac{v_r(t)-v_l(t)}{L}[rad/seg]$$

$$R(t)=\frac{L}{2}\frac{v_r(t)+v_l(t)}{v_r(t)-v_l(t)}$$

La velocidad tangencial del vehículo se calcula entonces como:

$$v(t)=\omega(t)R(t)=\frac{v_r(t)+v_l(t)}{2}[m/s]$$

Las velocidades tangenciales de las ruedas son:

$$v_l(t)=r\omega_l(t)$$

$$v_r(t)=r\omega_r(t)$$

Donde $\omega_l(t)$ y $\omega_r(t)$ son las velocidades angulares izquierda y derecha de las ruedas
alrededor de sus ejes, respectivamente.

Teniendo en cuenta las relaciones anteriores, la cinemática del robot (en coordenadas locales) se puede expresar como:

\begin{equation*}
\begin{bmatrix}
vx_m(t) \\
vy_m(t) \\
\omega(t)
\end{bmatrix}=
\begin{bmatrix}
\frac{1}{2} & \frac{1}{2} \\
0 & 0 \\
\frac{1}{L} & -\frac{1}{L}
\end{bmatrix}
\begin{bmatrix}
v_r(t) \\
v_l(t) \\
\end{bmatrix}
\label{eq:2} \tag{2}
\end{equation*}

\begin{equation*}
\begin{bmatrix}
v_r(t) \\
v_l(t) \\
\end{bmatrix}=
\begin{bmatrix}

\end{bmatrix}

\end{bmatrix}
\label{eq:3} \tag{3}
\end{equation*}

\end{bmatrix}
\label{eq:Cinematica_interna} \tag{5}
\end{equation*}

De la ecuacion $\eqref{eq:Cinematica_interna}$, se puede notar que el robot movil no tiene $vy_m(t)$, ademas que $vy_m(t)=v(t)$, con estas consideraciones optenemos:

\begin{equation*}
\begin{bmatrix}
v(t) \\
\omega(t)
\end{bmatrix}=
\begin{bmatrix}
\frac{r}{2} & \frac{r}{2} \\
\frac{r}{L} & -\frac{r}{L}

\end{equation*}

La cinemática externa del robot (en coordenadas globales) está dada por:

\begin{equation*}
\begin{bmatrix}

\dot\varphi(t)
\end{bmatrix}=
\begin{bmatrix}

\begin{equation*}
\mathbf{\dot{\eta}}=\mathbf{J}\mathbf{v}
\label{eq:9} \tag{9}
\end{equation*}

##  Implementación del Modelo del Robot Movil

### Importar las librerias necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
import time
import matplotlib as mpl
```

#### Tiempos para la simulacion

```python
t_sample=0.1
t_final=10+t_sample
t=np.arange(0,t_final,t_sample)
t=t.reshape(1,t.shape[0])
```

### Parametros del robot diferencial 
#### Radio de las ruedas


```python
r=0.195/2
```

#### Distancia entre cada rueda

```python
L=0.381
```

### Cinematica 
Se inicializa los vectores vacios para que sean almacenados en cada instante de tiempo

```python
xp=np.zeros((t.shape[0],t.shape[1]))
yp=np.zeros((t.shape[0],t.shape[1]))
```

### Velocidades Generales
Se inicializa los vectores vacios para que sean almacenados en cada instante de tiempo

```python
u=1*np.ones((t.shape[0],t.shape[1]))
w=0.5*np.ones((t.shape[0],t.shape[1]))
```

### Velocidades Angulares cada Rueda
Se inicializa los vectores vacios para que sean almacenados en cada instante de tiempo

```python
w_r=0*np.ones((t.shape[0],t.shape[1]))
w_l=0*np.ones((t.shape[0],t.shape[1]))
```

### Integracion
Se iniciliza los valores para que sean almacenados en cada instante de tiempo

```python
x=np.zeros((t.shape[0],t.shape[1]+1))
y=np.zeros((t.shape[0],t.shape[1]+1))
phi=np.zeros((t.shape[0],t.shape[1]+1))
```

### Condiciones iniciales del sistema

```python
x[0,0]=0
y[0,0]=0
phi[0,0]=(0*np.pi)/180
```

### Funcion del modelo del robot
En esta seccion se aplica la formula (7)

```python
def Jacobiano(v,phi):
    J=np.matrix([[np.cos(phi),0],[np.sin(phi),0]])
    hp=J*v
    return hp
```

### Integracion numerica del sistema
En esta seccion de aplica la formula (8)

```python
def euler(z,zp,t_sample):
    z=z+zp*t_sample
    return z
```

### Conversion de velocidades Generales a cada rueda
En esta seccion se utiliza la formula (6) 

```python
def conversion(v,r,L):
    T=np.matrix([[r/2,r/2],[r/L,-r/L]])
    tranformacion_ruedas=np.linalg.inv(T)*v
    return tranformacion_ruedas[0,0],tranformacion_ruedas[1,0]
```

### Bucle simulacion

```python
for k in range(0,t.shape[1]):

    ## Vector de velocidades del sistema
    v=np.array([[u[0,k]],[w[0,k]]])
    
    ## conversion de velocidades generales a velocidades angulares de cada rueda
    w_r[0,k],w_l[0,k]=conversion(v,r,L)
    
    ## Velocidad del punto modelado
    hp=Jacobiano(v,phi[0,k])
    
    ## Obtener las velocidades en Cada eje
    xp[0,k]=hp[0,0]
    yp[0,k]=hp[1,0]
    
    ## Integracion numerica
    x[0,k+1]=euler(x[0,k],xp[0,k],t_sample)
    y[0,k+1]=euler(y[0,k],yp[0,k],t_sample)
    phi[0,k+1]=euler(phi[0,k],w[0,k],t_sample)
    
```

### Funcion para las graficas del sistema

```python
def grafica(sty,titulo,x,y,etiqueta,ejex,ejey,color):
    mpl.style.use(sty)
    fig, ax = plt.subplots()
    ax.set_title(titulo.format(sty), color='0')
    ax.set_xlabel(ejex)
    ax.set_ylabel(ejey)
    ax.plot(x, y, color, label=etiqueta)
    ax.grid(linestyle='--', linewidth='0.3', color='black')
    legend = ax.legend(loc='upper right', shadow=False, fontsize='small')
    
```

### Graficas del sistema
#### Trayectoria del robot

```python
grafica('default','Trayectoria',x[0,:],y[0,:],'$\mathbf{\eta(t)}$','$x[m]$','$y[m]$','b')
```

#### Velocidades generales del robot

```python
grafica('default','Velocidad Lineal',t[0,:],u[0,:],'$\mu(t)$','$t[s]$','$[m/s]$','g')
grafica('default','Velocidad Angular',t[0,:],w[0,:],'$\omega(t)$','$t[s]$','$[rad/s]$','r')
```

#### Velocidades de cada rueda del robot

```python
grafica('default','Velocidad angular rueda derecha',t[0,:],w_r[0,:],'$\omega_{r}(t)$','$t[s]$','$[rad/s]$','r')
grafica('default','Velocidad angular rueda Izquierda',t[0,:],w_l[0,:],'$\omega_{l}(t)$','$t[s]$','$[rad/s]$','b')
```
